```python
import random
from numpy import array

def randNum(x:int, y:int):
    return random.randrange(x, y)
```

Generate data for the Rules Based AutoTables algorithm (Transpose Operation)

```python
# Creates a list containing set to ""
width, heigth = 5, 4
transpose_dataset = [["" for x in range(width)] for y in range(heigth)]
labels = ("Dollar\t", "Year\t", "Name\t", "City\t")

# Set first row to dollar value
for i in range(width):
    transpose_dataset[0][i] = "$" + str(randNum(1, 1000))

# Set second row to year value
for i in range(width):
    transpose_dataset[1][i] = str(randNum(1950, 2000))

# Set third row to name
names = ("John", "George", "Nicole", "Rachel", "Alex", "Lauren", "Paul", "Kim")
for i in range(width):
    transpose_dataset[2][i] = names[randNum(0, len(names))]

# Set fourth row to City
cities = ("Dallas", "Austin", "Houston", "San Antonio", "Fort Worth", "Lubbock", "Galveston")
for i in range(width):
    transpose_dataset[3][i] = cities[randNum(0, len(cities))]


print("\tOptionA\tOptionB\tOptionC\tOptionD\tOptionE")
for i in range(heigth):
    print(labels[i], end="")
    print(transpose_dataset[i])
```

Generate data for the Rules Based AutoTables algorithm (Pivot Operation)

```python
# Creates a list containing set to ""
heigth = 20
attributeNum = 4
counter = 0
pivot_dataset = ["" for i in range(heigth)]
labels = ("Dollar", "Year", "Name", "City")
names = ("John", "George", "Nicole", "Rachel", "Alex", "Lauren", "Paul", "Kim")
cities = ("Dallas", "Austin", "Houston", "San Antonio", "Fort Worth", "Lubbock", "Galveston")

# Create Pivot List
for i in range(int(heigth / attributeNum)):

    # Add dollar value
    pivot_dataset[counter] = "$" + str(randNum(1, 1000))

    # Add year value
    pivot_dataset[counter+1] = str(randNum(1950, 2000))

    # Add name value
    pivot_dataset[counter+2] = names[randNum(0, len(names))]

    # Add city value
    pivot_dataset[counter+3] = cities[randNum(0, len(cities))]

    counter += 4


options = ("OptionA","OptionB","OptionC","OptionD","OptionE")
for i in range(heigth):
    if i % attributeNum == 0:
        print("-------------- " + options[int(i / attributeNum)] + " --------------")
        
    # print(i, end=" ")
    print(labels[i%attributeNum], end="\t")
    print(pivot_dataset[i])
```

Generate data for the Rules Based AutoTables algorithm (Wide to Long Operation)

```python
# Create a list containing ""
width, heigth = 9, 5
wide_to_long_dataset = [["" for x in range(width)] for y in range(heigth)]

options = ("OptionA\t","OptionB\t","OptionC\t","OptionD\t","OptionE\t")
year_label = ("2018", "2019", "2020")
cat_label = ("Revenue", "Sold", "Margin")


# Create wide to long table
for i in range(heigth):
    for j in range(len(cat_label)):
        wide_to_long_dataset[i][3*j] = "$" + str(randNum(1, 10000))         # Create dollar value
        wide_to_long_dataset[i][(3*j)+1] = str(randNum(1, 1000))            # Create number value
        wide_to_long_dataset[i][(3*j)+2] = str(randNum(1, 50)) + "%"        # Create percentage value




# Print header
print("\t", end="")
for i in year_label:
    for j in cat_label:
        print(i + " " + j, end="   ")
print("")

# Print data
for i in range(heigth):
    print(options[i], end="")
    print(wide_to_long_dataset[i])

```

Algorithms to transform data

```python
def tranpose(data):
    dataShape = array(data).shape
    transformedData = [["" for x in range(dataShape[0])] for y in range(dataShape[1])]

    # Transform Data
    colNum = 0
    for rowNum in range(len(data[0])):
        transformedData[rowNum] = [data[0][rowNum], data[1][rowNum], data[2][rowNum], data[3][rowNum]]

    # Print table
    print("\t" + str(labels))
    for i in range(len(transformedData)):
        print(options[i], end="  ")
        print(transformedData[i])



def pivot(data):
    dataShape = len(data)
    transformedData = [["" for x in range(attributeNum)] for y in range(len(options))]

    # Transform Data
    for i in range(dataShape):
        transformedData[int(i/4)][i%4] = data[i]

    # Print table
    print("\t" + str(labels))
    for i in range(len(transformedData)):
        print(options[i], end="  ")
        print(transformedData[i])



def widetolong(data):
    
    dataShape = array(data).shape
    rowNum = len(year_label) * len(options)
    transformedData = [["" for x in range(len(cat_label))] for y in range(rowNum)]


    # Transform Data
    counter = 0
    for row in data:
        transformedData[counter+0] = [row[0], row[1], row[2]]
        transformedData[counter+1] = [row[3], row[4], row[5]]
        transformedData[counter+2] = [row[6], row[7], row[8]]
        counter += 3



    # Print table
    rowCounter = 0
    print("\t\t" + str(cat_label))
    print("-"*45)
    for i in range(len(options)):
        for j in range(len(year_label)):
            print(options[i] + "  " + year_label[j], end="   ")
            print(transformedData[rowCounter])
            rowCounter += 1
        print("-"*45)


# tranpose(transpose_dataset)
# pivot(pivot_dataset)
# widetolong(wide_to_long_dataset)
```

Rules Based AutoTables Algorithm

```python
def autotables(dataset):

    isPivot:bool = False
    isTranpose:bool = True
    isWideToLong:bool = False
    shape = array(dataset).shape


    # Check for pivot table (false until true)
    try:
        type(shape[1])
    except:
        shape = [shape[0], 1]
        isPivot = True
        isTranpose = False

    # Check for transpose (true until false)
    if isPivot is False:
        rowCounter = 0
        for row in dataset:
            for item in row:
                # Check first row
                if rowCounter == 0 and item[0] != "$":
                    isTranpose = False
                    break

                # Check second row
                if rowCounter == 1 and item[0] != "1" and item[1] != "9":
                    print(item[0])
                    print(item[1])
                    isTranpose = False
                    break

                # Check third row
                if rowCounter == 2 and item not in names:
                    isTranpose = False
                    break

                # Check fourth row
                if rowCounter == 3 and item not in cities:
                    isTranpose = False
                    break


            if isTranpose is False:
                break

            rowCounter += 1

    # Check if data is wide to long
    if isPivot is False and isTranpose is False:
        isWideToLong = True


    # Transform table
    if isPivot is True and isTranpose is False and isWideToLong is False:
        print("Pivot detected\n")
        pivot(dataset)
    elif isPivot is False and isTranpose is True and isWideToLong is False:
        print("Tranpose detected\n")
        tranpose(dataset)
    elif isPivot is False and isTranpose is False and isWideToLong is True:
        print("Wide to Long detected\n")
        widetolong(dataset)
    else:
        print(isPivot, end=" ")
        print(isTranpose, end=" ")
        print(isWideToLong)


# Used for testing 
# autotables(pivot_dataset)
# autotables(transpose_dataset)
# autotables(wide_to_long_dataset)

```
